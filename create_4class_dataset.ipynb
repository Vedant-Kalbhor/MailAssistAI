```python
import pandas as pd
import re

# Load your uploaded dataset
df = pd.read_csv("spam(1).csv", encoding="latin1")

# Keep only required columns
df = df[["v1", "v2"]]

df.columns = ["label", "text"]

df.head()
```

```python
def categorize_ham(text):

    text = text.lower()

    # FOLLOW UP RULES
    follow_up_keywords = [
        "please respond",
        "waiting for your reply",
        "kindly reply",
        "reminder",
        "follow up",
        "as discussed",
        "let me know",
        "awaiting response"
    ]

    for k in follow_up_keywords:
        if k in text:
            return "follow_up"

    # IMPORTANT RULES
    important_keywords = [
        "meeting",
        "deadline",
        "urgent",
        "asap",
        "important",
        "schedule",
        "appointment",
        "project",
        "report",
        "submission",
        "tomorrow",
        "today"
    ]

    for k in important_keywords:
        if k in text:
            return "important"

    # Default normal
    return "normal"
```

```python
def convert_label(row):

    if row["label"] == "spam":
        return "spam"

    # For ham emails apply rules
    return categorize_ham(row["text"])
```

```python
df["new_label"] = df.apply(convert_label, axis=1)

df.head()
```

```python
df["new_label"].value_counts()
```

```python
final_df = df[["text", "new_label"]]

final_df.to_csv("processed_4class_emails.csv", index=False)

print("Saved new dataset!")
```
